```python
#ゲーム実行
agent_list=[]
count=0
for i in range(100):
    count+=1
    agent_list.append("age"+str(count))
number_agent1=int(input())
number_agent2=int(input())
number_agent3=int(input())
total=number_agent1+number_agent2+number_agent3+1
print(number_agent1,number_agent2,number_agent3)
for i in range(number_agent1):
    agent_list[i+1]=agent1()
for i in range(number_agent2):
    agent_list[number_agent1+i+1]=agent2()
for i in range(number_agent3):
    agent_list[number_agent1+number_agent2+i+1]=agent3()
agent_list[0]=main_agent()
print(agent_list)
```

```python
check_list=[20]
select_list=[]
for i in range(100):
    check_list.append(check_list[-1]+20)
won_number_list10=[]
for freq in check_list:
  times=freq
  won=[1]
  won_agent=[]
  won_dict={}
  num_list=list(range(1,101))
  won_agent_dict={"agent1":0,"agent2":0,"agent3":0,"main_agent":0}
  for q in range(100):
    won_dict[q+1]=0
  for reset_dict in range(times):
    game_dict={}
    agent_dict={}
    for game in range(total):
      a=agent_list[game].game()
      if a in game_dict:
          game_dict[a]+=1
      else:
          game_dict[a]=1
          if game==0:
             agent_dict[a]="main_agent"
          elif game+1<=number_agent1:
            agent_dict[a]="agent1"
          elif game+1<=number_agent1+number_agent2:
             agent_dict[a]="agent2"
          elif game+1<=total:
             agent_dict[a]="agent3"
    for check in range(total):
      if check in game_dict and game_dict[check]==1:
          won.append(check)
          won_agent_dict[agent_dict[check]]+=1
          won_dict[check]+=1
          break
  won=won[1:]
  won_number_list10.append(won_agent_dict["main_agent"]/times)

print(won_number_list10)
print(select_list)
```
